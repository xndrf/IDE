```python
import matplotlib.pyplot as plt

year = [1950, 1975, 2000, 2018]
population = [2.12, 3.681, 5.312, 6.981]

plt.plot(year, population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population')
plt.show()
```

```python
plt.scatter(year, population)
plt.show()
```
